```python
# import pandas as pd
# import numpy as np
# import scipy.stats as stats
# import re
# pd.options.display.max_rows = 200

# nhl_df=pd.read_csv("assets/nhl.csv") # Load csv
# nhl_df = nhl_df[nhl_df['year'] == 2018] # Keep only results for 2018
# rows_toDrop = [0,9,18,26] # Dropping rows that are not needed - Divisions
# nhl_df.drop(rows_toDrop, inplace = True) # Dropping rows that are not needed - Divisions
# nhl_df = nhl_df.reset_index(drop=True) # Resetting index after dropping rows
# nhl_df[["W", "L"]] = nhl_df[["W", "L"]].apply(pd.to_numeric) # Changing the type of values for W and L to numeric
# #print(type(nhl_df['W'][0]))
# nhl_df['Win_loss_Ratio'] = nhl_df["W"]/ (nhl_df["W"] + nhl_df["L"]) # Creating win_loss_column
# nhl_df['Team'] = nhl_df['team'].str.replace(r"\*","") # Removing the asterisk from the end of some team names
# nhl_df['Team'] = nhl_df['Team'].str.strip() # Removing any possible spaces in team names
# nhl_df = nhl_df[['Team', 'Win_loss_Ratio']] # Keep only 'Team' and 'Win_loss_Ratio' columns
# nhl_df['Metropolitan area'] = nhl_df['Team'] # Create the Metropolitan area column
# nhl_df['NHL Team'] = nhl_df['Team'].str.split().str[-1] # Spliting team names to NHL Team column
# nhl_df['NHL Team'] = nhl_df['NHL Team'].replace(["Leafs", "Wings", "Jackets", "Knights"], # Fixing names
#                                           ["Maple Leafs","Red Wings","Blue Jackets", "Golden Knights"])
# # ***** Best Here ***** Removing NHL Teams names from Metropolitan area column
# nhl_df['Metropolitan area'] = nhl_df['Metropolitan area'].replace(nhl_df['NHL Team'],'', regex = True)
# nhl_df = nhl_df.astype({'Metropolitan area': str,'NHL Team': str, 'Win_loss_Ratio': float})
# nhl_df['Metropolitan area'] = nhl_df['Metropolitan area'].str.strip()

# nhl_df = nhl_df[['NHL Team', 'Win_loss_Ratio']] # Keeping certain columns
# #print(nhl_df)

# print('==============================================================================')
# #======================================================================================================================

# cities=pd.read_html("assets/wikipedia_data.html")[1]
# cities=cities.iloc[:-1,[0,3,5,6,7,8]] # Here we choose specific columns to keep and also we discard the last row.
# cities.rename(columns={"Population (2016 est.)[8]": "Population", 'NHL' : 'NHL Team'},inplace=True) # Rename population
# #cities['Population'] = cities['Population'].apply(pd.to_numeric) # Changing the type of values for 'Population (2016 est.)[8]' to numeric

# cities['NHL Team']= cities['NHL Team'].str.replace('\[.*\]',"")
# cities['NHL Team']= cities['NHL Team'].str.strip()
# cities = cities[['Metropolitan area', 'NHL Team', 'Population']]
# cities = cities.astype({'Metropolitan area': str,'NHL Team': str, 'Population': float})
# #print(cities)

# cities= cities.replace(['','—'], np.NaN).dropna()
# cities = cities.reset_index(drop=True)
# #print(cities)

# # This part here splits the concatenated rows see: (RangersIslandersDevils), into individual rows
# cities = (cities.set_index(['Metropolitan area', 'Population'])
# .apply(lambda x: x.str.split('([A-Z][a-z]*)([A-Z][a-z]*)')
# .explode())
# .reset_index()
# .replace("",np.nan)
# .dropna()
# .reset_index())
# #print(cities)
# #======================================================================================================================
# merged = pd.merge(cities, nhl_df,  how ='outer', on = 'NHL Team')
# merged = merged.groupby('Metropolitan area').agg({'Win_loss_Ratio': np.mean, 'Population': np.mean})
# print(merged)



def nhl_correlation(): 
    # YOUR CODE HERE
    
    import pandas as pd
    import numpy as np
    import scipy.stats as stats
    import re

    nhl_df=pd.read_csv("assets/nhl.csv") # Load csv
    nhl_df = nhl_df[nhl_df['year'] == 2018] # Keep only results for 2018
    rows_toDrop = [0,9,18,26] # Dropping rows that are not needed - Divisions
    nhl_df.drop(rows_toDrop, inplace = True) # Dropping rows that are not needed - Divisions
    nhl_df = nhl_df.reset_index(drop=True) # Resetting index after dropping rows
    nhl_df[["W", "L"]] = nhl_df[["W", "L"]].apply(pd.to_numeric) # Changing the type of values for W and L to numeric
    #print(type(nhl_df['W'][0]))
    nhl_df['Win_loss_Ratio'] = nhl_df["W"]/ (nhl_df["W"] + nhl_df["L"]) # Creating win_loss_column
    nhl_df['Team'] = nhl_df['team'].str.replace(r"\*","") # Removing the asterisk from the end of some team names
    nhl_df['Team'] = nhl_df['Team'].str.strip() # Removing any possible spaces in team names
    nhl_df = nhl_df[['Team', 'Win_loss_Ratio']] # Keep only 'Team' and 'Win_loss_Ratio' columns
    nhl_df['Metropolitan area'] = nhl_df['Team'] # Create the Metropolitan area column
    nhl_df['NHL Team'] = nhl_df['Team'].str.split().str[-1] # Spliting team names to NHL Team column
    nhl_df['NHL Team'] = nhl_df['NHL Team'].replace(["Leafs", "Wings", "Jackets", "Knights"], # Fixing names
                                              ["Maple Leafs","Red Wings","Blue Jackets", "Golden Knights"])
    # ***** Best Here ***** Removing NHL Teams names from Metropolitan area column
    nhl_df['Metropolitan area'] = nhl_df['Metropolitan area'].replace(nhl_df['NHL Team'],'', regex = True)
    nhl_df = nhl_df.astype({'Metropolitan area': str,'NHL Team': str, 'Win_loss_Ratio': float})
    nhl_df['Metropolitan area'] = nhl_df['Metropolitan area'].str.strip()

    nhl_df = nhl_df[['NHL Team', 'Win_loss_Ratio']] # Keeping certain columns
    #print(nhl_df)

    #print('==============================================================================')
    #======================================================================================================================

    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]] # Here we choose specific columns to keep and also we discard the last row.
    #print(cities)
    cities.rename(columns={"Population (2016 est.)[8]": "Population", 'NHL' : 'NHL Team'},inplace=True) # Rename population
    #cities['Population'] = cities['Population'].apply(pd.to_numeric) # Changing the type of values for 'Population (2016 est.)[8]' to numeric

    cities['NHL Team']= cities['NHL Team'].str.replace('\[.*\]',"")
    cities['NHL Team']= cities['NHL Team'].str.strip()
    cities = cities[['Metropolitan area', 'NHL Team', 'Population']]
    cities = cities.astype({'Metropolitan area': str,'NHL Team': str, 'Population': float})
    #print(cities)

    cities= cities.replace(['','—'], np.NaN).dropna()
    cities = cities.reset_index(drop=True)
    #print(cities)

    # This part here splits the concatenated rows see: (RangersIslandersDevils), into individual rows
    cities = (cities.set_index(['Metropolitan area', 'Population'])
    .apply(lambda x: x.str.split('([A-Z][a-z]*)([A-Z][a-z]*)')
    .explode())
    .reset_index()
    .replace("",np.nan)
    .dropna()
    .reset_index())
    #print(cities)
    #======================================================================================================================
    merged = pd.merge(cities, nhl_df,  how ='outer', on = 'NHL Team')
    merged = merged.groupby('Metropolitan area').agg({'Win_loss_Ratio': np.mean, 'Population': np.mean})
    merged = merged.astype({'Win_loss_Ratio': float,'Population': int})
    #print(merged)
    
    
    #raise NotImplementedError()
    population_by_region = merged['Population'] # pass in metropolitan area population from cities
    win_loss_by_region = merged['Win_loss_Ratio'] # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]
    
    assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return stats.pearsonr(population_by_region, win_loss_by_region)[0]

nhl_correlation()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re
pd.options.display.max_rows = 200


#========================================================================================================
# Cities
#========================================================================================================
def citiesSports(sport):
       
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    cities.rename(columns={"Population (2016 est.)[8]": "Population"},inplace=True) # Rename population
    #cities.drop('Population', axis =1, inplace = True)
    cities = cities.replace(to_replace='\[.*\]', value='', regex=True)
    cities = cities.applymap(lambda x: x.strip() if isinstance(x, str) else x) # This one strips away spaces in the whole dataframe

    data = cities[['Metropolitan area', 'Population', sport]]

    data= data.replace(['','—'], np.NaN).dropna()
    data = data.reset_index(drop=True)
    # The regex for this exercise was found here: https://ycchen00.github.io/2020/11/09/Intro2DS/Assignment4/
    df = (data.set_index(['Metropolitan area', 'Population'])
    .apply(lambda x: x.str.split('([A-Z]{0,2}[a-z0-9]*\ [A-Z]{0,2}[a-z0-9]*|[A-Z]{0,2}[a-z0-9]*)([A-Z]{0,2}[a-z0-9]*\ [A-Z]{0,2}[a-z0-9]*|[A-Z]{0,2}[a-z0-9]*)([A-Z]{0,2}[a-z0-9]*\ [A-Z]{0,2}[a-z0-9]*|[A-Z]{0,2}[a-z0-9]*)')
    .explode())
    .reset_index()
    .replace(['',' '],np.nan)
    .dropna()
    .reset_index())
    df = df.astype({'Metropolitan area': str,sport: str, 'Population': int})
    df.drop('index', axis =1, inplace = True)
    return  df

NFL_df = citiesSports('NFL')
#print(NFL_df)
NBA_df = citiesSports('NBA')
#print(NBA_df)
NHL_df = citiesSports('NHL')
#print(NHL_df)
MLB_df = citiesSports('MLB')
#print(MLB_df)

#========================================================================================================
# MLB Teams
# =======================================================================================================
mlb_df=pd.read_csv("assets/mlb.csv")
mlb_df = mlb_df[mlb_df['year'] == 2018]
mlb_df = mlb_df[['team', 'W-L%']] # Keep only 'Team' and 'Win_loss_Ratio' columns
mlb_df['Metropolitan area'] = mlb_df['team'] # Create the Metropolitan area column
mlb_df['MLB'] = mlb_df['team'].str.split().str[-1] # Spliting team names to NHL Team column

# Had to do these manually becase there were 2 Sox and couldnt automate
mlb_df['MLB'][0]='Red Sox'
mlb_df['MLB'][8]='White Sox'
mlb_df['MLB'][3]='Blue Jays'

# ***** Best Here ***** Removing NHL Teams names from Metropolitan area column
mlb_df['Metropolitan area'] = mlb_df['Metropolitan area'].replace(mlb_df['MLB'],'', regex = True)
mlb_df = mlb_df.astype({'Metropolitan area': str,'MLB': str, 'W-L%': float})
mlb_df['Metropolitan area'] = mlb_df['Metropolitan area'].str.strip()
mlb_df = mlb_df[['MLB', 'W-L%']]
#print(mlb_df)

MLB_merge = pd.merge(MLB_df, mlb_df,  how ='outer', on = 'MLB')
MLB_merge = MLB_merge.groupby('Metropolitan area').agg({'W-L%': np.nanmean, 'Population': np.nanmean})
MLB_merge = MLB_merge.astype({'W-L%': float,'Population': int})
MLB = MLB_merge[['W-L%']]
#print(MLB)

#========================================================================================================
# NHL Teams
# =======================================================================================================
nhl_df=pd.read_csv("assets/nhl.csv")
nhl_df = nhl_df[nhl_df['year'] == 2018] # Keep only results for 2018
rows_toDrop = [0,9,18,26] # Dropping rows that are not needed - Divisions
nhl_df.drop(rows_toDrop, inplace = True) # Dropping rows that are not needed - Divisions
nhl_df = nhl_df.reset_index(drop=True) # Resetting index after dropping rows
nhl_df[["W", "L"]] = nhl_df[["W", "L"]].apply(pd.to_numeric) # Changing the type of values for W and L to numeric
#print(type(nhl_df['W'][0]))
nhl_df['W-L%'] = nhl_df["W"]/ (nhl_df["W"] + nhl_df["L"]) # Creating win_loss_column
nhl_df['Team'] = nhl_df['team'].str.replace(r"\*","") # Removing the asterisk from the end of some team names
nhl_df['Team'] = nhl_df['Team'].str.strip() # Removing any possible spaces in team names
nhl_df = nhl_df[['Team', 'W-L%']] # Keep only 'Team' and 'Win_loss_Ratio' columns
nhl_df['Metropolitan area'] = nhl_df['Team'] # Create the Metropolitan area column
nhl_df['NHL'] = nhl_df['Team'].str.split().str[-1] # Spliting team names to NHL Team column
nhl_df['NHL'] = nhl_df['NHL'].replace(["Leafs", "Wings", "Jackets", "Knights"], # Fixing names
                                              ["Maple Leafs","Red Wings","Blue Jackets", "Golden Knights"])
# ***** Best Here ***** Removing NHL Teams names from Metropolitan area column
nhl_df['Metropolitan area'] = nhl_df['Metropolitan area'].replace(nhl_df['NHL'],'', regex = True)
nhl_df = nhl_df.astype({'Metropolitan area': str,'NHL': str, 'W-L%': float})
nhl_df['Metropolitan area'] = nhl_df['Metropolitan area'].str.strip()

nhl_df = nhl_df[['NHL', 'W-L%']] # Keeping certain columns
#print(nhl_df)

NHL_merge = pd.merge(NHL_df, nhl_df,  how ='outer', on = 'NHL')
NHL_merge = NHL_merge.groupby('Metropolitan area').agg({'W-L%': np.nanmean, 'Population': np.nanmean})
NHL_merge = NHL_merge.astype({'W-L%': float,'Population': int})
NHL = NHL_merge[['W-L%']]
#print(NHL)

#========================================================================================================
# NBA Teams
# =======================================================================================================
nba_df=pd.read_csv("assets/nba.csv")
nba_df = nba_df[~nba_df['team'].str.contains("Division")] # Dropping rows that are not needed - Divisions
nba_df = nba_df[nba_df['year'] == 2018]
nba_df = nba_df.reset_index(drop=True) # Resetting index after dropping rows
nba_df['team'] = nba_df['team'].str.replace(r"\*","") # Removing the asterisk from the end of some team names
nba_df['team']= nba_df['team'].str.replace('\(.*\)',"")
nba_df['team'] = nba_df['team'].str.strip() # Removing any possible spaces in team names
nba_df = nba_df[['team', 'W/L%']] # Keep only 'Team' and 'Win_loss_Ratio' columns
nba_df.rename(columns = {'W/L%' : 'W-L%'}, inplace = True)
#nba_df = nba_df.rename({'W/L%' : 'W-L%'}, axis=1)
nba_df['Metropolitan area'] = nba_df['team'] # Create the Metropolitan area column
nba_df['NBA'] = nba_df['team'].str.split().str[-1] # Spliting team names to NHL Team column
nba_df['NBA'] = nba_df['NBA'].replace(["Blazers",],["Trail Blazers"])

# ***** Best Here ***** Removing NHL Teams names from Metropolitan area column
nba_df['Metropolitan area'] = nba_df['Metropolitan area'].replace(nba_df['NBA'],'', regex = True)
nba_df = nba_df.astype({'Metropolitan area': str,'NBA': str, 'W-L%': float})
nba_df['Metropolitan area'] = nba_df['Metropolitan area'].str.strip()
nba_df = nba_df[['NBA', 'W-L%']]

#print(nba_df)

NBA_merge = pd.merge(NBA_df, nba_df,  how ='outer', on = 'NBA')
NBA_merge = NBA_merge.groupby('Metropolitan area').agg({'W-L%': np.nanmean, 'Population': np.nanmean})
NBA_merge = NBA_merge.astype({'W-L%': float,'Population': int})
NBA = NBA_merge[['W-L%']]
#print(NBA)

#========================================================================================================
# NFL Teams
# =======================================================================================================
nfl_df=pd.read_csv("assets/nfl.csv")
nfl_df = nfl_df[nfl_df['year'] == 2018]
nfl_df = nfl_df[['team', 'W-L%']] # Keep only 'Team' and 'Win_loss_Ratio' columns
nfl_df = nfl_df[~nfl_df['team'].str.contains("AFC")] # Dropping rows that are not needed
nfl_df = nfl_df[~nfl_df['team'].str.contains("NFC")] # Dropping rows that are not needed
nfl_df['team'] = nfl_df['team'].str.replace(r"\*","") # Removing the asterisk from the end of some team names
nfl_df['team']= nfl_df['team'].str.replace('\+',"")
#print(nfl_df)
nfl_df['team'] = nfl_df['team'].str.strip() # Removing any possible spaces in team names
nfl_df['Metropolitan area'] = nfl_df['team'] # Create the Metropolitan area column
nfl_df['NFL'] = nfl_df['team'].str.split().str[-1] # Spliting team names to NHL Team column
# ***** Best Here ***** Removing NHL Teams names from Metropolitan area column
nfl_df['Metropolitan area'] = nfl_df['Metropolitan area'].replace(nfl_df['NFL'],'', regex = True)
nfl_df = nfl_df.astype({'Metropolitan area': str,'NFL': str, 'W-L%': float})
nfl_df['Metropolitan area'] = nfl_df['Metropolitan area'].str.strip()
nfl_df = nfl_df[['NFL', 'W-L%']]

#print(nfl_df)

NFL_merge = pd.merge(NFL_df, nfl_df,  how ='outer', on = 'NFL')
NFL_merge = NFL_merge.groupby('Metropolitan area').agg({'W-L%': np.nanmean, 'Population': np.nanmean})
NFL_merge = NFL_merge.astype({'W-L%': float,'Population': int})
NFL = NFL_merge[['W-L%']]
#print(NFL)
    
#========================================================================================================
#////////////////////////////////////////////////////////////////////////////////////////////////////////
#========================================================================================================

def what_sport(sport):
    if sport == 'MLB':
        return MLB
    elif sport == 'NHL':
        return NHL
    elif sport == 'NBA':
        return NBA
    elif sport == 'NFL':
        return NFL
    else:
        print('Enter valid sport!')
    
        
def sports_team_performance():
    # YOUR CODE HERE
    #raise NotImplementedError()
    
    # Note: p_values is a full dataframe, so df.loc["NFL","NBA"] should be the same as df.loc["NBA","NFL"] and
    # df.loc["NFL","NFL"] should return np.nan
    sports = ['NFL', 'NBA', 'NHL', 'MLB']
    p_values = pd.DataFrame({k:np.nan for k in sports}, index=sports)
    
    for sport in sports:
        for sport_2 in sports:
            if sport != sport_2:
                Dataframe = pd.merge(what_sport(sport), what_sport(sport_2), how = 'inner', on = 'Metropolitan area')
                #print('Sports:', sport, sport_2)
                #print(Dataframe)
                p_values.loc[sport, sport_2]=stats.ttest_rel(Dataframe['W-L%_x'],Dataframe['W-L%_y'])[1]
                #print(p_values)
    
    
    assert abs(p_values.loc["NBA", "NHL"] - 0.02) <= 1e-2, "The NBA-NHL p-value should be around 0.02"
    assert abs(p_values.loc["MLB", "NFL"] - 0.80) <= 1e-2, "The MLB-NFL p-value should be around 0.80"
    return p_values

sports_team_performance()
```
